```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Generate Original Kite Network and Nodes Names
G=nx.krackhardt_kite_graph()
nx.draw(G,with_labels=True)
plt.show()
```

```python
# Rename All Nodes using Dictionary
mapping = {0: "Ben", 1: "Ana", 2: "Dave", 3: "John", 4: "Carl", 5: "Sam",6: "Sue",7: "Jess",8: "Ella",9: "Eaye"}
G = nx.relabel_nodes(G, mapping)

# Print the Newly Named Graph
nx.draw(G,with_labels=True)
plt.show()
```

```python
# Print All Centrality 
print("Betweenness")
b = nx.betweenness_centrality(G)
for v in G.nodes():
 print("%s: %5.3f" % (v, b[v]))
print("Degree centrality")
d = nx.degree_centrality(G)
for v in G.nodes():
 print("%s: %5.3f" % (v, d[v]))
print("Closeness centrality")
c = nx.closeness_centrality(G)
for v in G.nodes():
 print("%s: %5.3f" % (v, c[v]))
print("Eigenvector centrality")
centrality = nx.eigenvector_centrality(G)
A=sorted((v, '{:0.2f}'.format(c)) for v, c in centrality.items())
for xs in A:
  print(": ".join(map(str, xs)))
```
